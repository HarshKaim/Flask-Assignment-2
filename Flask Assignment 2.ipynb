Q1. Explain GET and POST methods.

Ans. GET is used to retrieve data from a server. Data is passed as a part of the URL. It does not alter the server's state.

POST method is used to send data to the server. Data is passes as a part of the request. It is more secure. It does alter the server's state.

Q2. Why is request used in Flask?

Ans. Request object is used to handle incoming HTTP requests. It helps in accessing requests' details made by client.

Q3. Why is redirect() used in Flask?

Ans. redirect() method is used to redirect the user to a different URL automatically. For e.g. redirecting a user to the home page after a successful form submission

Q4. What are templates in Flask? Why is the render_template() function used?

Ans. A template in Flask is HTML code with placeholders in it. The value in these placeholders is assigned when we use render_template() method and load the HTML web page to the user with all the actual values created after execution of the backend code which is written in python (or in some other language).
